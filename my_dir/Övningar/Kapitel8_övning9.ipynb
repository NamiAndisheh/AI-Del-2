```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

# Ladda ResNet50 modellen (förtränad på ImageNet)
print("Laddar ResNet50 modell...")
model = ResNet50(weights='imagenet')
print("Modell laddad!")
```

```python
# Ange sökvägen till din bild här
# Exempel: "my_dir/Kod_från_läraren/pictures/lejon.jpg"
image_path = "my_dir/Kod_från_läraren/pictures/lejon.jpg"

# Eller ange din egen bild:
# image_path = "sökväg/till/din/bild.jpg"
```

```python
# Ladda och visa bilden
if os.path.exists(image_path):
    # Öppna bilden
    img = Image.open(image_path)
    
    # Visa bilden
    plt.figure(figsize=(10, 8))
    plt.imshow(img)
    plt.axis('off')
    plt.title('Bilden som ska analyseras', fontsize=16)
    plt.show()
    
    # Förbered bilden för modellen (ResNet50 kräver 224x224)
    img_resized = img.resize((224, 224))
    x = image.img_to_array(img_resized)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    # Klassificera bilden
    print("Analyserar bilden...")
    preds = model.predict(x, verbose=0)
    resultat = decode_predictions(preds, top=3)[0]
    
    # Visa resultatet
    print("\n" + "="*50)
    print("Jag tror det är en...")
    print("="*50)
    for i, (id, label, prob) in enumerate(resultat):
        print(f"{i+1}. {label} ({prob*100:.2f}% sannolikhet)")
    
else:
    print(f"Kunde inte hitta bilden på sökvägen: {image_path}")
    print("\nTillgängliga exempelbilder:")
    example_dir = "my_dir/Kod_från_läraren/pictures"
    if os.path.exists(example_dir):
        for file in os.listdir(example_dir):
            if file.endswith(('.jpg', '.jpeg', '.png')):
                print(f"  - {example_dir}/{file}")
```
